Step 2: Extracting Links

Parse the HTML using BeautifulSoup and extract links using a CSS selector:

```python
soup = BeautifulSoup(html, "html.parser")
links = [a['href'] for a in soup.select("a[href]")]
```

Step 3: Crawling Multiple Pages

Extend the logic to visit every page, storing visited URLs to prevent duplicates:



```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = "https://www.scrapingcourse.com/ecommerce/"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

links = set()

for a in soup.select("a[href]"):
    link = a['href']
    full_url = urljoin(url, link)

    if full_url.startswith("http"):
        links.add(full_url)

# Print all links
for link in links:
    print(link)
```

Step 4: Extracting Product Data

Enhance the script to scrape product information, such as name, price, and ratings, from each page:



```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

base_url = "https://www.scrapingcourse.com/ecommerce/"
response = requests.get(base_url)
soup = BeautifulSoup(response.content, "html.parser")

clean_links = set()

for a in soup.select("a[href]"):
    link = a['href']
    full_url = urljoin(base_url, link)

    if (
        "ecommerce/product/" in full_url and
        "#" not in full_url and
        "?" not in full_url and
        full_url.endswith("/") and
        full_url.startswith("https://www.scrapingcourse.com")
    ):
        clean_links.add(full_url)

print("Total product links:", len(clean_links))

products = []

for link in clean_links:
    try:
        print("Scraping:", link)

        res = requests.get(link, timeout=5)
        page = BeautifulSoup(res.content, "html.parser")
        
        name_tag = page.select_one("h1")
        
        price_tag = page.select_one(".price")

        if name_tag and price_tag:
            products.append({
                "name": name_tag.text.strip(),
                "price": price_tag.text.strip(),
                "url": link
            })

    except Exception as e:
        print("Error:", e)

print("Total products:", len(products))
print(products[:5])
```

Step 5: Exporting Data to CSV

```python
import csv

with open("products.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["name", "price", "url"])
    
    writer.writeheader()
    writer.writerows(products)
```
