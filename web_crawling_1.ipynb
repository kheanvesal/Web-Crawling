1. Basic Web Crawling

This example demonstrates a simple way to crawl a webpage. A GET request is made to a website using the requests library, and the HTTP status code along with the HTML content of the response is printed.

```python
import requests

url = "https://www.w3schools.com/"
response = requests.get(url)

if response.status_code == 200:
    print("HTML Content:")
    print(response.text)
else:
    print(f"Failed to fetch the page. Status Code: {response.status_code}")
```

2. Crawling JSON Data

Fetch JSON data from APIs and convert it into Python dictionaries. For example, a GET request is sent to the ISS location API to retrieve and display the current position of the ISS.

```python
import requests

url = "http://api.open-notify.org/iss-now.json"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print("ISS Location Data:", data)
else:
    print(f"Error fetching data. Status Code: {response.status_code}")
```

3. Web Crawling Images



```python
import requests

image_url = "https://infatica.io/blog/content/images/2020/06/an-extensive-overview-of-python-web-crawlers.png"
response = requests.get(image_url)

if response.status_code == 200:
    with open("downloaded_image.png", "wb") as file:
        file.write(response.content)
    print("Image saved as downloaded_image.png")
else:
    print("Failed to download the image.")
```

4. Crawling Elements Using XPath

Extract specific elements from web pages using XPath. Here's an example of extracting the current temperature from a weather website.

```python
import requests
from lxml import etree

url = "https://www.timeanddate.com/weather/india/noida"

headers = {
    "User-Agent": "Mozilla/5.0"
}

try:
    response = requests.get(url, headers=headers, timeout=10)

    if response.status_code == 200:
        tree = etree.HTML(response.content)
        temp = tree.xpath("//div[contains(@class,'h2')]/text()")
        
        if temp:
            print("Temperature:", temp[0].strip())
        else:
            print("Temperature not found")

    else:
        print("Status Code:", response.status_code)

except requests.exceptions.RequestException as e:
    print("Error:", e)
```

5. Crawling a Web Page for Most Frequent Words

Analyze the content of a webpage to determine the most frequently used words.



```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
import re

url = "https://www.bbc.com/news/world"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    text = soup.get_text()
    words = re.findall(r'\b\w+\b', text.lower())
    word_counts = Counter(words).most_common(10)
    print("Most Frequent Words:", word_counts)
else:
    print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
```
